```python
q1 = df["frp"].quantile(0.60)
q2 = df["frp"].quantile(0.90)
```

Severity Func

```python
def frp_to_severity(frp):
  if frp <= q1:
    return "Low"
  elif frp <= q2:
    return "Medium"
  else:
    return "High"
```

```python
df["severity"] = df["frp"].apply(frp_to_severity)
```

```python
df["severity"].value_counts()
```

```python
df.head()
```

```python
df["daynight"] = df["daynight"].map({"D": 1, "N": 0})
```

```python
df["brightness_diff"] = df["brightness"] - df["bright_t31"]
```

```python
le = LabelEncoder()

df['severity'] = le.fit_transform(df['severity'])
df['confidence'] = le.fit_transform(df['confidence'])
```

```python
df = df.sample(n=150000, random_state=42)
```

Model Training

```python
x = df[[
    "brightness",
    "bright_t31",
    "brightness_diff",
    "scan",
    "track",
    "latitude",
    "longitude",
    "confidence"
]]
y = df["severity"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = RandomForestClassifier(
    n_estimators=200,
    max_depth = 25,
    class_weight="balanced",
    random_state=42,
    n_jobs = -1,
    min_samples_leaf=5,
)
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
print(classification_report(y_test, pred))
```
